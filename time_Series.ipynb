resample ()  VID 87


````python
import pandas as pd

col = ["date" , "close" , "high" , "low" , "close"]
apple_stock = pd.read_csv("AAPL.csv" , usecols=col , parse_dates=["date"], index_col = "date")
apple_stock.head(5)
````

````python
month = apple_stock.resample("MS") # <-- MS is frequency
month
````

````python
month.groups ## will give indices (indexes) at the end
````

````python
month.get_group("2019-11-01 00:00:00+0000") # give exact timestamp
# give entire moyn excluding last day and weekends
````

````python
month.mean()
month.mean().round(2)
````

````python
month2 = apple_stock.resample("M" ,  kind="period")
month2
````

````python
month2.groups
````

````python
month2.mean().round(2)
month2.max().round(2)
````

````python
# agg() it aggregate
    # "low" is column
month2.agg({"low" :["min" , "max"] , "high" : ["min" , "max"]})
````
